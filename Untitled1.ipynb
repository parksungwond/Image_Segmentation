```python
#조건문
#점수를 입력받아 등급을 출력하세요
#등급표 81~100 A / 60~80 B / 41~60 C / 21~40 D /0~20 E

score = int(input("점수를 입력하세요:"))

if score > 80:
  re = "A"
elif score > 60:
  re = "B"
elif score > 40:
  re = "C"
elif score > 20:
  re = "D"
else:
  re = "E"

print('점수는 %d점이며' %score, '등급은 %s등급 입니다.'%re) 
```

```python
# 함수를 만들고 두 수를 입력 받아 연산하세요

a = int(input('첫 번째 정수를 입력하세요:'))
b = int(input('두 번째 정수를 입력하세요:'))

print("덧셈 함수 연산: ", a+b)
print("곱셈 함수 연산: ", a*b)
print("제곱 함수 연산: ", a**b)
```

```python
#알고리즘 생성
#피보나치 수열을 계산하는 함수를 생성하고 자연수 10을 넣어 실행하시오

def pivo(n: int) -> int:
  if n >= 2:
    return pivo(n - 2) + pivo(n - 1)
  else:
      return n

      print(pivo(10))
```
